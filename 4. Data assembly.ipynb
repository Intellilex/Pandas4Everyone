# 4.4 Merging multiple datasets

```python
person = pd.read_csv('./files/survey_person.csv')
site = pd.read_csv('./files/survey_site.csv')
survey = pd.read_csv('./files/survey_survey.csv')
visited = pd.read_csv('./files/survey_visited.csv')
```

## 4.4.1 one-to-one merge

```python
visited_subset = visited.loc[[0, 2, 6]]

o2o_merge = site.merge(visited_subset, left_on='name', right_on='site')

o2o_merge
```

### 4.4.2 Many-to-one merge

```python
m2o_merge = site.merge(visited, left_on='name', right_on='site')
m2o_merge
```

## 4.4.3 Many-to-many merge

```python
ps = person.merge(survey, left_on='ident', right_on='person')
vs = visited.merge(survey, left_on='ident', right_on='taken')
ps, vs
```

```python
ps_vs = ps.merge(vs,
                 left_on=['ident', 'taken', 'quant', 'reading'],
                right_on=['person', 'ident', 'quant', 'reading'])

ps_vs.loc[0]
```
